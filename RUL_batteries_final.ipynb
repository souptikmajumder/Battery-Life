```python
#TimeStamp,Voltage,Current,Ah,Wh,Power,Battery_Temp_degC,Time,Chamber_Temp_degC = ([],)*9
TimeStamp=[];Voltage = [];Current = [];Ah = [];Wh=[];Power=[];Battery_Temp_degC=[];
Time=[];Chamber_Temp_degC=[];

TimeStamp_f,Voltage_f,Current_f,Ah_f,Wh_f,Power_f,Battery_Temp_degC_f,Time_f,Chamber_Temp_degC_f = ([],)*9

from itertools import chain

for i in range(0,25):
    TimeStamp.append(list(chain.from_iterable(deg_25_list[i]['meas'][0][0][0])))
    Voltage.append(list(chain.from_iterable(deg_25_list[i]['meas'][0][0][1])))
    Current.append(list(chain.from_iterable(deg_25_list[i]['meas'][0][0][2])))  
    Ah.append(list(chain.from_iterable(deg_25_list[i]['meas'][0][0][3])))        
    Wh.append(list(chain.from_iterable(deg_25_list[i]['meas'][0][0][4])))
    Power.append(list(chain.from_iterable(deg_25_list[i]['meas'][0][0][5])))
    Battery_Temp_degC.append(list(chain.from_iterable(deg_25_list[i]['meas'][0][0][6])))
    Time.append(list(chain.from_iterable(deg_25_list[i]['meas'][0][0][7])))
    Chamber_Temp_degC.append(list(chain.from_iterable(deg_25_list[i]['meas'][0][0][8])))


#unzipping all the list elements in the final lists to form dataframe
TimeStamp_f= list(chain.from_iterable(TimeStamp))
Voltage_f = list(chain.from_iterable(Voltage))
Current_f = list(chain.from_iterable(Current))
Ah_f = list(chain.from_iterable(Ah))
Wh_f = list(chain.from_iterable(Wh))
Power_f = list(chain.from_iterable(Power))
Battery_Temp_degC_f = list(chain.from_iterable(Battery_Temp_degC))
Time_f = list(chain.from_iterable(Time))
Chamber_Temp_degC_f = list(chain.from_iterable(Chamber_Temp_degC))


final_timestamp = []
final_timestamp = list(chain.from_iterable(TimeStamp_f))


#create dataframe for charge and discharge cycles at 25 degree
df_1 = pd.DataFrame(list(zip(final_timestamp,Voltage_f,Current_f,Ah_f,Wh_f,Power_f,Battery_Temp_degC_f,Time_f,Chamber_Temp_degC_f)),columns=['TimeStamp','Voltage','Current',
                                                               'Ah','Wh','Power','Battery_Temp_degC','Time',
                                                               'Chamber_Temp_degC'])
df_1['TimeStamp']=df_1['TimeStamp'].apply(lambda x : pd.to_datetime(str(x)))
df_1 = df_1.sort_values(by='TimeStamp',ascending=True)
```


##### As observed 28.4 mOhms is the minimum reistance at full load capacity of the battery. So as the charge discharge cycles continue the value of the Impedance will increase at each full load capacity. So to calculate the State of Health we will take the percentage increase in resistances at full load capacity and from there the RUL can be calculated within 70-80 % of the SOH.

```python
#calculating State of Health (SOH)
impedance_change_percent = []
soh = [] #list of state of health
count_a = 0;
count_b = 0;
min_impedance = df_drives_25deg['Impedance'].min();

for i in range(len(df_drives_25deg['discharge'])):
    if(df_drives_25deg['discharge'][i] == 1):
        count_a = (df_drives_25deg['Impedance'][i] - min_impedance)
        impedance_change_percent.append(count_a/min_impedance * 100)
    else:
        impedance_change_percent.append(count_a/min_impedance * 100)

df_drives_25deg['Impedance_change'] = impedance_change_percent
#df_drives_25deg['Impedance_change'] = round(df_drives_25deg['Impedance_change'],2)
df_drives_25deg['SOH'] = 100.0 - (df_drives_25deg['Impedance_change'])
```

```python
# imports
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
import numpy
from tensorflow.keras.optimizers import Adam
import keras
from matplotlib import pyplot
from keras.callbacks import EarlyStopping
import pandas as pd 
from sklearn.preprocessing import LabelEncoder
# Create model
model = Sequential()
model.add(Dense(128, activation="relu", input_dim=3))
model.add(Dense(64, activation="relu"))
model.add(Dense(32, activation="relu"))
model.add(Dense(16, activation="relu"))
model.add(Dense(8, activation="relu"))
model.add(Dense(4, activation="relu"))
# Since the regression is performed, a Dense layer containing a single neuron with a linear activation function.
# Typically ReLu-based activation are used but since it is performed regression, it is needed a linear activation.
model.add(Dense(1, activation="linear"))

# Compile model: The model is initialized with the Adam optimizer and then it is compiled.
model.compile(loss='mean_squared_error', optimizer=Adam(lr=1e-3, decay=1e-3 / 100))
```

```python
# Patient early stopping
es = EarlyStopping(monitor='val_loss', mode='min', verbose=1, patience=100)

# Fit the model
history = model.fit(X_train, y_train, validation_data=(X_valid, y_valid), epochs=1000, batch_size=100, verbose=2, callbacks=[es])

# Calculate predictions
PredTrainSet = model.predict(X_train)
PredValSet = model.predict(X_valid)


# Save predictions
numpy.savetxt("trainresults.csv", PredTrainSet, delimiter=",")
numpy.savetxt("valresults.csv", PredValSet, delimiter=",")
```

```python
# Plot training history
pyplot.plot(history.history['loss'], label='train')
pyplot.plot(history.history['val_loss'], label='test')
plt.title('Training Vs Validation loss')
pyplot.legend()
pyplot.show()
```
